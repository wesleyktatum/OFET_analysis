```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_75 = pd.read_csv("75c_box_plot.csv") 
data_100 = pd.read_csv("100c_box_plot.csv")
data_125 = pd.read_csv("125c_box_plot.csv")
data_150 = pd.read_csv("150c_box_plot.csv")
```

```python
Temperature_75 = data_75.Temperature
Time_75 = data_75.Time
mu_lin_avg_75 = data_75.mu_lin_avg
mu_lin_sd_75 = data_75.mu_lin_sd
r_lin_avg_75 = data_75.r_lin_avg
r_lin_sd_75 = data_75.r_lin_sd
on_off_avg_75 = data_75.on_off_avg
on_off_sd_75 = data_75.on_off_sd
vt_avg_75 = data_75.vt_avg
vt_sd_75 = data_75.vt_sd
```

```python
all_min_75_mu_lin = [mu_lin_avg_75[0:5], mu_lin_avg_75[6:9], 
           mu_lin_avg_75[10:13], mu_lin_avg_75[14:21]]
bp_75_mu_lin = plt.boxplot(all_min_75_mu_lin, labels=("5", "10", "15", "30"))
plt.title('75째C mu_lin_avg')
plt.xlabel('Minutes')
plt.ylabel('mu_lin_avg')
plt.savefig("bp_mu_lin_avg_75")
```

```python
all_min_75_r_lin = [r_lin_avg_75[0:5], r_lin_avg_75[6:9],
                 r_lin_avg_75[10:13], r_lin_avg_75[14:21]]
bp_75_r_lin = plt.boxplot(all_min_75_r_lin, labels=("5", "10", "15", "30"))
plt.title('75째C r_lin_avg')
plt.xlabel('Minutes')
plt.ylabel('r_lin_avg')
plt.savefig("bp_r_lin_avg_75")
```

```python
all_min_75_on_off = [on_off_avg_75[0:5], on_off_avg_75[6:9],
                 on_off_avg_75[10:13], on_off_avg_75[14:21]]
bp_75_on_off = plt.boxplot(all_min_75_on_off, labels=("5", "10", "15", "30"))
plt.title('75째C on_off_avg')
plt.xlabel('Minutes')
plt.ylabel('on_off_avg')
plt.savefig("bp_on_off_avg_75")
```

```python
all_min_75_vt = [vt_avg_75[0:5], vt_avg_75[6:9],
                 vt_avg_75[10:13], vt_avg_75[14:21]]
bp_75_vt = plt.boxplot(all_min_75_vt, labels=("5", "10", "15", "30"))
plt.title('75째C vt_avg')
plt.xlabel('Minutes')
plt.ylabel('vt_avg')
plt.savefig("bp_vt_avg_75")
```
